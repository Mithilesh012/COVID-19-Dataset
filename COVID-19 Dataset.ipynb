We have taken a small dataset of Covid-19, just for your understaning purpose. You have to work on the original dataset which contains about 19000 rows.

The data used here is till 29-April-2020 and has records as on 29-April-2020.

This data is available as a CSV file, downloaded from Kaggle.

We will analyze this data using the Pandas DataFrame.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(r"covid_19_data.csv")
```

```python
data
```

```python
data.count()
# Null Values means Missing Values
```

```python
data.isnull().sum()
```

```python
sns.heatmap(data.isnull())
plt.show()
```

```python
# Convert 'Date' to datetime
data['Date'] = pd.to_datetime(data['Date'])

# Set the style for seaborn
sns.set(style='whitegrid')
```

```python
# Create a bar plot for Deaths
plt.figure(figsize=(13, 6))
sns.barplot(data=data, x='State', y='Deaths', ci=None)
plt.title('Total Deaths by State')
plt.xlabel('State')
plt.ylabel('Total Deaths')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# Create a bar plot for Recovered
plt.figure(figsize=(13, 6))
sns.barplot(data=data, x='State', y='Recovered', ci=None)
plt.title('Total Recovered Cases by State')
plt.xlabel('State')
plt.ylabel('Total Recovered')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
